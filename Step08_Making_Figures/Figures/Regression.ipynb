```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


data = pd.read_csv(r'/Users/xujiahua/Desktop/Twins/Draw_2020_0612/178G_Final.csv') # 数据长度

val='lamy1'

font = {'weight': 'normal', 'size':8}

fig = plt.figure()
ax1 = fig.add_subplot(111)
sns.regplot(x="MN2", y=val,data=data ,scatter_kws={'alpha':1,'s':1,'color':'#808080'},line_kws=dict(color='#FF4500', lw=2),ax=ax1);
#sns.regplot(x="Age1", y=val,data=datafemaleC ,scatter_kws={'alpha':1,'s':5,'color':'#FA8072'},line_kws=dict(color='#FA8072', lw=2),ax=ax1);
#sns.set_style("whitegrid", {"axes.grid": False})
sns.despine(offset=None, trim=False, left=False, bottom=False)
plt.tick_params(labelsize=8)
plt.ylabel('Connectivity (a.u.)',font)
ax1.yaxis.tick_left()
ax1.yaxis.set_label_position("left")
plt.tick_params(labelsize=8)
plt.xlabel('Maternal negative parenting',font)
plt.locator_params('y',nbins=4)
ax1=plt.gca()
ax1.spines['bottom'].set_linewidth(1)
ax1.spines['left'].set_linewidth(1)
fig = plt.gcf()      
#plt.xticks( np.array([7,8,9,10,11,12]), ('7','8','9','10','11','12') )
ax1.yaxis.tick_left()
ax1.yaxis.set_label_position("left")
fig.set_size_inches(2,2)
fig.savefig("/Users/xujiahua/Desktop/Twins/Draw_2020_0612/"+val+".png", dpi=600, transparent=False,bbox_inches = 'tight')

```
